<h3>Read data from CSV and create a DF</h3>

```python
import pandas as pd
df = pd.read_csv('data.csv')
df.shape
```

Filtering the Age over 40

```python
df[df['Age'] > 40].head()
```

Adding the value to float function

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

Create a new df to find the best value player

```python
df1 = pd.DataFrame(df, columns=['Name', 'Wage', 'Value'])

wage = df1['Wage'].replace(['\€'], '', regex=True).apply(value_to_float)
value = df1['Value'].replace(['\€'], '', regex=True).apply(value_to_float)

df1['Wage'] = wage
df1['Value'] = value

df1['Diff'] = df1['Value'] - df1['Wage']
df1.sort_values('Diff', ascending=False).head()
```

<h3>Chart the result with plotly

```python
import plotly.express as px
fig = px.scatter(df1, x='Wage', y='Value', hover_data=['Name', 'Diff'])
fig.show()
```
